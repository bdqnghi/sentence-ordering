```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
from scipy import stats
from itertools import permutations
import torch
import random
```

```python
kts = []
for _ in range(10000):
    
    perm = torch.randperm(6).tolist()
    
    kt, _ = stats.kendalltau(range(6), perm)
    kts.append(kt)
```

```python
plt.hist(kts)
```

```python
kts = []
for _ in range(10):
    
    perms = permutations(list(range(6)))
    
    for perm in perms:
        kt, _ = stats.kendalltau(range(6), perm)
        kts.append(kt)
```

```python
plt.hist(kts)
```
