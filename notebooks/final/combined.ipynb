```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
from sorder.metrics import Metrics
```

```python
lr_metrics = Metrics.from_file('linear_regression/gp.json')
```

```python
cnn_metrics = Metrics.from_file('neural_regression/cnn.json')
```

```python
cbow_metrics = Metrics.from_file('neural_regression/cbow.json')
```

```python
lstm_metrics = Metrics.from_file('neural_regression/lstm.json')
```

```python
pairs_metrics = Metrics.from_file('pairs/gp.360.json')
```

```python
pick_next_metrics = Metrics.from_file('pick_next/gp.200.json')
```

```python
ctx_reg_metrics = Metrics.from_file('context_regression/gp.205.json')
```

# Perfect match

```python
lr_xy = lr_metrics.perfect_order_pct_by_len()
lr_x, lr_y = zip(*lr_xy.items())

cnn_xy = cnn_metrics.perfect_order_pct_by_len()
cnn_x, cnn_y = zip(*cnn_xy.items())

cbow_xy = cbow_metrics.perfect_order_pct_by_len()
cbow_x, cbow_y = zip(*cbow_xy.items())

lstm_xy = lstm_metrics.perfect_order_pct_by_len()
lstm_x, lstm_y = zip(*lstm_xy.items())

pairs_xy = pairs_metrics.perfect_order_pct_by_len()
pairs_x, pairs_y = zip(*pairs_xy.items())

pick_next_xy = pick_next_metrics.perfect_order_pct_by_len()
pick_next_x, pick_next_y = zip(*pick_next_xy.items())

ctx_reg_xy = ctx_reg_metrics.perfect_order_pct_by_len()
ctx_reg_x, ctx_reg_y = zip(*ctx_reg_xy.items())

plt.figure(figsize=(10, 6))
plt.ylim(-0.05, 1)
plt.xticks(ctx_reg_x)

plt.xlabel('Sentence length')
plt.ylabel('Perfect Match Ratio')

plt.title('Perfect Match Ratio')
plt.plot(lr_x, lr_y, marker='D', label='Linear regression')
plt.plot(cbow_x, cbow_y, marker='s', label='CBOW regression')
plt.plot(cnn_x, cnn_y, marker='o', label='CNN regression')
plt.plot(lstm_x, lstm_y, marker='h', label='LSTM regression')
plt.plot(pairs_x, pairs_y, marker='v', label='Pairwise ranking model')
plt.plot(pick_next_x, pick_next_y, marker='^', label='First/Next')
plt.plot(ctx_reg_x, ctx_reg_y, marker='x', label='Regression with context')
plt.legend(loc='upper right')
```
