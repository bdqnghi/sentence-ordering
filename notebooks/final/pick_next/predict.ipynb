# Pick next

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
import ujson

from sorder.metrics import Metrics
```

```python
with open('gp.200.json') as fh:
    metrics = Metrics(ujson.load(fh))
```

# Perfect order

```python
metrics.overall_perfect_order_pct()
```

```python
xy = metrics.perfect_order_pct_by_len()

x, y = zip(*xy.items())

plt.ylim(-0.05, 1)
plt.xticks(x)
plt.xlabel('Sentence length')
plt.ylabel('Percentage correct')
plt.plot(x, y)
```

# Kendall's tau
