```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
from scipy import stats
from itertools import permutations
import torch
import random
```

```python
kts = []
for _ in range(10000):
    
    perm = torch.randperm(6).tolist()
    
    kt, _ = stats.kendalltau(range(6), perm)
    kts.append(kt)
```

```python
plt.hist(kts)
```

```python
kts = []
for _ in range(10):
    
    perms = permutations(list(range(6)))
    
    for perm in perms:
        kt, _ = stats.kendalltau(range(6), perm)
        kts.append(kt)
```

```python
plt.hist(kts)
```

```python
kts = []
for _ in range(1000):
    seq = list(range(10))
    for _ in range(1):
        i1, i2 = random.sample(range(10), 2)
        seq[i1], seq[i2] = seq[i2], seq[i1]
        kt, _ = stats.kendalltau(range(10), seq)
        kts.append(kt)
```

```python
plt.hist(kts)
```

```python
kts = []
seq = list(range(20))
for _ in range(100):
    i = random.choice(range(0, len(seq)-2))
    seq[i], seq[i+1] = seq[i+1], seq[i]
    kt, _ = stats.kendalltau(range(len(seq)), seq)
    kts.append(kt)
```

```python
plt.plot(kts)
```

```python
plt.hist(kts)
```

```python
kts = []
for _ in range(1000):
    seq = list(range(10))
    i1, i2 = random.sample(range(10), 2)
    seq[i1], seq[i2] = seq[i2], seq[i1]
    kt, _ = stats.spearmanr(range(10), seq)
    kts.append(kt)
```

```python
plt.plot(kts)
```

```python
plt.hist(kts)
```

```python
kts = []
for _ in range(1000):
    
    perm = torch.randperm(20).tolist()
    
    kt, _ = stats.weightedtau(range(20), perm)
    kts.append(kt)
```

```python
plt.hist(kts)
```
